````python
# import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LinearRegression   # not ideal for binary, but kept for reference

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
````

````python
# Set random seeds for reproducibility
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
tf.random.set_seed(42)

# Load data

df = pd.read_csv(r'/content/swat_cloudburst_large_3100.csv')
print("Data shape:", df.shape)
df.head()

# =============================================
# 2. Exploratory Data Analysis & Preprocessing
# =============================================
# Check missing values
print("\nMissing values:\n", df.isnull().sum())

# Fill missing values (e.g., with median - more robust than 22)
df.fillna(df.median(numeric_only=True), inplace=True)

# Alternatively, if you want to use 22 as before:
# df.fillna(22, inplace=True)

# Check data types
df.info()

# Assume target column is 'Cloudburst_Event'
target_col = 'Cloudburst_Event'
feature_cols = ['Max_Daily_Rainfall_mm', 'Pressure_hPa', 'Humidity_%', 'Cloud_Cover_%']

# Verify columns exist
for col in feature_cols + [target_col]:
    if col not in df.columns:
        raise ValueError(f"Column '{col}' not found in CSV. Available: {df.columns.tolist()}")

# Separate features and target
X = df[feature_cols]
y = df[target_col]

# Check class distribution
print("\nClass distribution:\n", y.value_counts(normalize=True))

# Scale features (important for neural networks and logistic regression)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split (80-20)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)

print(f"Training set size: {X_train.shape[0]}")
print(f"Test set size: {X_test.shape[0]}")

# =============================================
# 3. Traditional Machine Learning Models
# =============================================
results = {}

# ----- Logistic Regression -----
lr = LogisticRegression(max_iter=1000, random_state=42)
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)
results['Logistic Regression'] = {
    'Accuracy': accuracy_score(y_test, y_pred_lr),
    'Precision': precision_score(y_test, y_pred_lr),
    'Recall': recall_score(y_test, y_pred_lr),
    'F1': f1_score(y_test, y_pred_lr)
}
print("Logistic Regression:")
print(classification_report(y_test, y_pred_lr))

# ----- Random Forest Classifier -----
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
results['Random Forest'] = {
    'Accuracy': accuracy_score(y_test, y_pred_rf),
    'Precision': precision_score(y_test, y_pred_rf),
    'Recall': recall_score(y_test, y_pred_rf),
    'F1': f1_score(y_test, y_pred_rf)
}
print("\nRandom Forest:")
print(classification_report(y_test, y_pred_rf))

# (Optional) Linear Regression – not suitable, but shown for completeness
# lr_reg = LinearRegression()
# lr_reg.fit(X_train, y_train)
# y_pred_reg = (lr_reg.predict(X_test) > 0.5).astype(int)

# =============================================
# 4. Deep Learning: LSTM Model
# =============================================
# LSTM expects 3D input: (samples, timesteps, features)
# Since each sample is independent, we set timesteps = 1
X_train_lstm = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
X_test_lstm = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))

# Build LSTM model
model_lstm = Sequential([
    LSTM(32, input_shape=(1, X_train.shape[1]), activation='tanh', return_sequences=False),
    Dropout(0.2),
    Dense(16, activation='relu'),
    Dropout(0.2),
    Dense(1, activation='sigmoid')
])

model_lstm.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model_lstm.summary()

# Early stopping to avoid overfitting
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Train the model
history = model_lstm.fit(
    X_train_lstm, y_train,
    validation_split=0.2,
    epochs=60,
    batch_size=16,
    callbacks=[early_stop],
    verbose=1
)

# Evaluate LSTM
y_pred_lstm_proba = model_lstm.predict(X_test_lstm)
y_pred_lstm = (y_pred_lstm_proba > 0.5).astype(int).ravel()

results['LSTM'] = {
    'Accuracy': accuracy_score(y_test, y_pred_lstm),
    'Precision': precision_score(y_test, y_pred_lstm),
    'Recall': recall_score(y_test, y_pred_lstm),
    'F1': f1_score(y_test, y_pred_lstm)
}
print("\nLSTM:")
print(classification_report(y_test, y_pred_lstm))

# =============================================
# 5. Model Comparison & Visualization
# =============================================
# Create comparison DataFrame
comparison_df = pd.DataFrame(results).T
print("\n=== Model Performance Comparison ===")
print(comparison_df.round(4))

# Bar plot of metrics
comparison_df.plot(kind='bar', figsize=(10,6), colormap='viridis')
plt.title('ML vs DL: Performance Comparison')
plt.ylabel('Score')
plt.ylim(0,1)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Confusion matrices
fig, axes = plt.subplots(1, 3, figsize=(15,4))
for i, (name, preds) in enumerate(zip(['Logistic Regression', 'Random Forest', 'LSTM'],
                                       [y_pred_lr, y_pred_rf, y_pred_lstm])):
    cm = confusion_matrix(y_test, preds)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[i])
    axes[i].set_title(f'{name}\nAccuracy: {results[name]["Accuracy"]:.3f}')
    axes[i].set_xlabel('Predicted')
    axes[i].set_ylabel('Actual')
plt.tight_layout()
plt.show()

# =============================================
# 6. Interactive Prediction with Best Model
# =============================================
# Choose the best model based on F1-score
best_model_name = max(results, key=lambda x: results[x]['F1'])
print(f"\nBest model based on F1-score: {best_model_name}")

if best_model_name == 'Logistic Regression':
    best_model = lr
elif best_model_name == 'Random Forest':
    best_model = rf
else:
    best_model = model_lstm

# Prediction function (handles scaling and LSTM reshaping)
def predict_cloudburst(rainfall, pressure, humidity, cloud_cover):
    # Create input array
    input_data = np.array([[rainfall, pressure, humidity, cloud_cover]])
    # Scale using the same scaler
    input_scaled = scaler.transform(input_data)
    if best_model_name == 'LSTM':
        input_scaled = input_scaled.reshape((1, 1, 4))
        proba = best_model.predict(input_scaled, verbose=0)[0][0]
    else:
        proba = best_model.predict_proba(input_scaled)[0][1] if hasattr(best_model, 'predict_proba') else best_model.predict(input_scaled)[0]
    prediction = 1 if proba >= 0.5 else 0
    return prediction, proba

# Interactive prediction
print("\n--- Cloudburst Prediction ---")
rainfall = float(input('Enter Max Daily Rainfall (mm): '))
pressure = float(input('Enter Pressure (hPa): '))
humidity = float(input('Enter Humidity (%): '))
cloud_cover = float(input('Enter Cloud Cover (%): '))

pred, prob = predict_cloudburst(rainfall, pressure, humidity, cloud_cover)
if pred == 1:
    print(f"\n⚠️ CLOUDBURST ALERT! (Probability: {prob:.2f})")
else:
    print(f"\n✅ No cloudburst predicted. (Probability: {prob:.2f})")
````

````python
import streamlit as st
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ML/DL Imports
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

# Page Layout configuration
st.set_page_config(page_title="Cloudburst Prediction Dashboard", layout="wide")

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# ---------------------------------------------
# 1. Cached Data Loading & Model Training Pipeline
# ---------------------------------------------
@st.cache_resource
def run_training_pipeline(csv_path):
    # Load and clean data
    df = pd.read_csv(csv_path)
    df.fillna(df.median(numeric_only=True), inplace=True)

    target_col = 'Cloudburst_Event'
    feature_cols = ['Max_Daily_Rainfall_mm', 'Pressure_hPa', 'Humidity_%', 'Cloud_Cover_%']

    X = df[feature_cols]
    y = df[target_col]

    # Scale features
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # Train-test split
    X_train, X_test, y_train, y_test = train_test_split(
        X_scaled, y, test_size=0.2, random_state=42, stratify=y
    )

    results = {}

    # --- Logistic Regression ---
    lr = LogisticRegression(max_iter=1000, random_state=42)
    lr.fit(X_train, y_train)
    y_pred_lr = lr.predict(X_test)
    results['Logistic Regression'] = {
        'Accuracy': accuracy_score(y_test, y_pred_lr),
        'Precision': precision_score(y_test, y_pred_lr),
        'Recall': recall_score(y_test, y_pred_lr),
        'F1': f1_score(y_test, y_pred_lr),
        'preds': y_pred_lr
    }

    # --- Random Forest ---
    rf = RandomForestClassifier(n_estimators=100, random_state=42)
    rf.fit(X_train, y_train)
    y_pred_rf = rf.predict(X_test)
    results['Random Forest'] = {
        'Accuracy': accuracy_score(y_test, y_pred_rf),
        'Precision': precision_score(y_test, y_pred_rf),
        'Recall': recall_score(y_test, y_pred_rf),
        'F1': f1_score(y_test, y_pred_rf),
        'preds': y_pred_rf
    }

    # --- LSTM ---
    X_train_lstm = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
    X_test_lstm = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))

    model_lstm = Sequential([
        LSTM(32, input_shape=(1, X_train.shape[1]), activation='tanh', return_sequences=False),
        Dropout(0.2),
        Dense(16, activation='relu'),
        Dropout(0.2),
        Dense(1, activation='sigmoid')
    ])
    model_lstm.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

    early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
    model_lstm.fit(
        X_train_lstm, y_train, validation_split=0.2,
        epochs=60, batch_size=16, callbacks=[early_stop], verbose=0
    )

    y_pred_lstm_proba = model_lstm.predict(X_test_lstm, verbose=0)
    y_pred_lstm = (y_pred_lstm_proba > 0.5).astype(int).ravel()

    results['LSTM'] = {
        'Accuracy': accuracy_score(y_test, y_pred_lstm),
        'Precision': precision_score(y_test, y_pred_lstm),
        'Recall': recall_score(y_test, y_pred_lstm),
        'F1': f1_score(y_test, y_pred_lstm),
        'preds': y_pred_lstm
    }

    # Determine best model based on F1-score
    best_model_name = max(results, key=lambda x: results[x]['F1'])
    if best_model_name == 'Logistic Regression':
        best_model = lr
    elif best_model_name == 'Random Forest':
        best_model = rf
    else:
        best_model = model_lstm

    return scaler, results, best_model, best_model_name, y_test

# ---------------------------------------------
# 2. Main Execution Flow
# ---------------------------------------------
st.title("⛈️ Cloudburst Prediction & Analytical System")
st.markdown("This dashboard trains ML/DL models on atmospheric datasets to predict severe localized cloudburst events.")

# Define hardcoded CSV path from your code snippet
DATA_PATH = '/content/swat_cloudburst_large_3100.csv'

try:
    with st.spinner("Processing dataset and training models (Logistic Regression, Random Forest, LSTM)..."):
        scaler, model_results, best_model, best_model_name, y_test = run_training_pipeline(DATA_PATH)
    st.success(f"Pipeline running successfully! Best model determined by F1-score: **{best_model_name}**")

    # Create clean comparison dictionary for presentation
    comparison_metrics = {k: {m: v[m] for m in ['Accuracy', 'Precision', 'Recall', 'F1']} for k, v in model_results.items()}
    comparison_df = pd.DataFrame(comparison_metrics).T

    # ---------------------------------------------
    # UI Tabs layout
    # ---------------------------------------------
    tab1, tab2 = st.tabs(["🔮 Real-time Prediction", "📊 Model Performance Analytics"])

    with tab1:
        st.header("Predictive Risk Parameters")
        st.markdown("Adjust the lower atmospheric variables below to run a real-time risk simulation.")

        col1, col2 = st.columns(2)
        with col1:
            rainfall = st.number_input('Max Daily Rainfall (mm)', min_value=0.0, max_value=500.0, value=12.0, step=1.0)
            pressure = st.number_input('Atmospheric Pressure (hPa)', min_value=900.0, max_value=1100.0, value=1008.0, step=1.0)
        with col2:
            humidity = st.number_input('Relative Humidity (%)', min_value=0.0, max_value=100.0, value=75.0, step=1.0)
            cloud_cover = st.number_input('Cloud Cover (%)', min_value=0.0, max_value=100.0, value=80.0, step=1.0)

        if st.button("Analyze Cloudburst Risk", type="primary"):
            # Shape and scale inputs
            input_data = np.array([[rainfall, pressure, humidity, cloud_cover]])
            input_scaled = scaler.transform(input_data)

            # Predict based on best model configuration
            if best_model_name == 'LSTM':
                input_scaled = input_scaled.reshape((1, 1, 4))
                prob = best_model.predict(input_scaled, verbose=0)[0][0]
            else:
                prob = best_model.predict_proba(input_scaled)[0][1] if hasattr(best_model, 'predict_proba') else best_model.predict(input_scaled)[0]

            pred = 1 if prob >= 0.5 else 0

            # Display localized Alert Card UI elements
            st.markdown("---")
            if pred == 1:
                st.error(f"### ⚠️ CLOUDBURST ALERT!\n\n**Calculated Probability:** `{prob:.2%}`\n\nAtmospheric levels indicate extreme structural hazard signatures. Evacuation protocols or warning indicators are advised for high-slope terrains.")
            else:
                st.success(f"### ✅ Atmospheric Environment Safe\n\n**Calculated Probability:** `{prob:.2%}`\n\nNo acute flash rain or cloudburst anomalies detected under current sensor inputs.")

    with tab2:
        st.header("Comparative Core Training Metrics")
        st.dataframe(comparison_df.style.highlight_max(axis=0, color='#2c5234'))

        # Plot 1: Performance Bar Metric Plot
        st.subheader("Classification Score Analysis Across Frameworks")
        fig, ax = plt.subplots(figsize=(9, 4.5))
        comparison_df.plot(kind='bar', ax=ax, color=['#4caf50', '#2196f3', '#ff9800', '#9c27b0'])
        ax.set_ylabel('Score Ratio')
        ax.set_ylim(0, 1.1)
        plt.xticks(rotation=0)
        ax.grid(axis='y', linestyle='--', alpha=0.4)
        st.pyplot(fig)

        # Plot 2: Joint Heatmap Confusion Matrices
        st.subheader("Confusion Matrices (Validation Split)")
        fig2, axes = plt.subplots(1, 3, figsize=(16, 4))
        for i, (name, metrics) in enumerate(model_results.items()):
            cm = confusion_matrix(y_test, metrics['preds'])
            sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[i], cbar=False)
            axes[i].set_title(f'{name}\nAccuracy: {metrics["Accuracy"]:.3f}')
            axes[i].set_xlabel('Predicted Classes')
            axes[i].set_ylabel('True Classes')
        plt.tight_layout()
        st.pyplot(fig2)

except FileNotFoundError:
    st.error(f"**Data File Not Found:** Could not trace the CSV source array situated at location: `{DATA_PATH}`. Please adapt your paths to continue.")
````

````python
%%writefile app.py
import streamlit as st
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

st.set_page_config(page_title="Cloudburst Prediction", layout="wide")
np.random.seed(42)
tf.random.set_seed(42)

@st.cache_resource
def run_training_pipeline(csv_path):
    df = pd.read_csv(r'/content/swat_cloudburst_large_3100.csv')
    df.fillna(df.median(numeric_only=True), inplace=True)
    X = df[['Max_Daily_Rainfall_mm', 'Pressure_hPa', 'Humidity_%', 'Cloud_Cover_%']]
    y = df['Cloudburst_Event']

    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)

    results = {}
    # Logistic Regression
    lr = LogisticRegression(max_iter=1000, random_state=42)
    lr.fit(X_train, y_train)
    y_pred_lr = lr.predict(X_test)
    results['Logistic Regression'] = {'Accuracy': accuracy_score(y_test, y_pred_lr), 'F1': f1_score(y_test, y_pred_lr), 'preds': y_pred_lr}

    # Random Forest
    rf = RandomForestClassifier(n_estimators=100, random_state=42)
    rf.fit(X_train, y_train)
    y_pred_rf = rf.predict(X_test)
    results['Random Forest'] = {'Accuracy': accuracy_score(y_test, y_pred_rf), 'F1': f1_score(y_test, y_pred_rf), 'preds': y_pred_rf}

    # LSTM
    X_train_lstm = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
    X_test_lstm = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))
    model_lstm = Sequential([
        LSTM(32, input_shape=(1, X_train.shape[1]), activation='tanh', return_sequences=False),
        Dropout(0.2),
        Dense(16, activation='relu'),
        Dropout(0.2),
        Dense(1, activation='sigmoid')
    ])
    model_lstm.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
    model_lstm.fit(X_train_lstm, y_train, validation_split=0.2, epochs=30, batch_size=16, callbacks=[early_stop], verbose=0)

    y_pred_lstm = (model_lstm.predict(X_test_lstm, verbose=0) > 0.5).astype(int).ravel()
    results['LSTM'] = {'Accuracy': accuracy_score(y_test, y_pred_lstm), 'F1': f1_score(y_test, y_pred_lstm), 'preds': y_pred_lstm}

    best_model_name = max(results, key=lambda x: results[x]['F1'])
    best_model = lr if best_model_name == 'Logistic Regression' else (rf if best_model_name == 'Random Forest' else model_lstm)
    return scaler, results, best_model, best_model_name, y_test

st.title("⛈️ Cloudburst Prediction System")
DATA_PATH = '/content/swat_cloudburst_large_3100.csv'

try:
    with st.spinner("Training models..."):
        scaler, model_results, best_model, best_model_name, y_test = run_training_pipeline(DATA_PATH)
    st.success(f"Best Engine: {best_model_name}")

    col1, col2 = st.columns(2)
    with col1:
        rainfall = st.number_input('Max Daily Rainfall (mm)', min_value=0.0, value=12.0)
        pressure = st.number_input('Atmospheric Pressure (hPa)', min_value=900.0, value=1008.0)
    with col2:
        humidity = st.number_input('Relative Humidity (%)', min_value=0.0, max_value=100.0, value=75.0)
        cloud_cover = st.number_input('Cloud Cover (%)', min_value=0.0, max_value=100.0, value=80.0)

    if st.button("Analyze Risk", type="primary"):
        input_data = np.array([[rainfall, pressure, humidity, cloud_cover]])
        input_scaled = scaler.transform(input_data)
        if best_model_name == 'LSTM':
            input_scaled = input_scaled.reshape((1, 1, 4))
            prob = best_model.predict(input_scaled, verbose=0)[0][0]
        else:
            prob = best_model.predict_proba(input_scaled)[0][1]

        st.markdown("---")
        if prob >= 0.5:
            st.error(f"### ⚠️ CLOUDBURST ALERT! (Probability: {prob:.2%})")
        else:
            st.success(f"### ✅ Environment Safe (Probability: {prob:.2%})")
except FileNotFoundError:
    st.error("Please upload your CSV to `/content/swat_cloudburst_large_3100.csv` first.")
````

````python
# 1. Initialize Virtual Display for Google Colab
!apt-get install -y xvfb x11-utils
!pip install pyvirtualdisplay
from pyvirtualdisplay import Display
import os

# Start virtual display buffer
disp = Display(visible=0, size=(800, 600))
disp.start()
os.environ["DISPLAY"] = f":{disp.display}"

# 2. Standard Imports
import tkinter as tk
from tkinter import ttk, messagebox
import numpy as np
import pandas as pd

# ML/DL Processing Imports
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score, accuracy_score

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

# Set random seeds
np.random.seed(42)
tf.random.set_seed(42)

# ---------------------------------------------
# 3. Background ML Pipeline Configuration
# ---------------------------------------------
DATA_PATH = '/content/swat_cloudburst_large_3100.csv'

def train_system():
    try:
        df = pd.read_csv(DATA_PATH)
        df.fillna(df.median(numeric_only=True), inplace=True)

        X = df[['Max_Daily_Rainfall_mm', 'Pressure_hPa', 'Humidity_%', 'Cloud_Cover_%']]
        y = df['Cloudburst_Event']

        scaler = StandardScaler()
        X_scaled = scaler.fit_transform(X)

        X_train, X_test, y_train, y_test = train_test_split(
            X_scaled, y, test_size=0.2, random_state=42, stratify=y
        )

        results = {}

        # Logistic Regression
        lr = LogisticRegression(max_iter=1000, random_state=42)
        lr.fit(X_train, y_train)
        results['Logistic Regression'] = {'model': lr, 'F1': f1_score(y_test, lr.predict(X_test))}

        # Random Forest
        rf = RandomForestClassifier(n_estimators=100, random_state=42)
        rf.fit(X_train, y_train)
        results['Random Forest'] = {'model': rf, 'F1': f1_score(y_test, rf.predict(X_test))}

        # LSTM
        X_train_lstm = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
        X_test_lstm = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))

        lstm = Sequential([
            LSTM(32, input_shape=(1, X_train.shape[1]), activation='tanh'),
            Dropout(0.2),
            Dense(16, activation='relu'),
            Dropout(0.2),
            Dense(1, activation='sigmoid')
        ])
        lstm.compile(optimizer='adam', loss='binary_crossentropy')
        early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
        lstm.fit(X_train_lstm, y_train, validation_split=0.2, epochs=20, batch_size=16, callbacks=[early_stop], verbose=0)

        y_pred_lstm = (lstm.predict(X_test_lstm, verbose=0) > 0.5).astype(int).ravel()
        results['LSTM'] = {'model': lstm, 'F1': f1_score(y_test, y_pred_lstm)}

        # Extract best engine
        best_name = max(results, key=lambda x: results[x]['F1'])
        return scaler, results[best_name]['model'], best_name, None
    except Exception as e:
        return None, None, None, str(e)

# Run Background Engine Setup
scaler, best_model, best_model_name, error_msg = train_system()

# ---------------------------------------------
# 4. Building the Tkinter Application Window
# ---------------------------------------------
root = tk.Tk()
root.title("Cloudburst Analytical Prediction Interface")
root.geometry("550x450")

# Apply a modern clean look layout style
style = ttk.Style(root)
style.theme_use('clam')

# Header Title Block
header = ttk.Label(root, text="⛈️ Cloudburst Detection Framework", font=("Helvetica", 16, "bold"), foreground="#1e3d59")
header.pack(pady=15)

if error_msg:
    err_lbl = ttk.Label(root, text=f"Pipeline Error: CSV Missing at /content/\n({error_msg})", foreground="red", font=("Helvetica", 11))
    err_lbl.pack(pady=20)
else:
    # Notification Banner of trained model
    engine_lbl = ttk.Label(root, text=f"Active Prediction Engine: {best_model_name}", font=("Helvetica", 10, "italic"), foreground="green")
    engine_lbl.pack(pady=5)

    # Core Data Form Container
    form_frame = ttk.LabelFrame(root, text=" Atmospheric Parameters Input ", padding=20)
    form_frame.pack(padx=20, pady=15, fill="both", expand=True)

    # Input placement mapping
    fields = [
        ("Max Daily Rainfall (mm):", "12.0"),
        ("Atmospheric Pressure (hPa):", "1008.0"),
        ("Relative Humidity (%):", "75.0"),
        ("Cloud Cover (%):", "80.0")
    ]

    entries = {}
    for label_text, default_val in fields:
        row = ttk.Frame(form_frame)
        row.pack(fill="x", pady=6)

        lbl = ttk.Label(row, text=label_text, width=30, anchor="w", font=("Helvetica", 10))
        lbl.pack(side="left")

        ent = ttk.Entry(row)
        ent.insert(0, default_val)
        ent.pack(side="right", expand=True, fill="x")
        entries[label_text] = ent

    # Display Result Window Field
    result_var = tk.StringVar()
    result_var.set("Adjust parameters and hit 'Evaluate Risk Profile'")
    result_lbl = ttk.Label(root, textvariable=result_var, font=("Helvetica", 11, "bold"), wraplength=450, justify="center")

    # Prediction Evaluation Execution Logic
    def run_prediction():
        try:
            # Parse parameters
            rf = float(entries["Max Daily Rainfall (mm):"].get())
            pr = float(entries["Atmospheric Pressure (hPa):"].get())
            hm = float(entries["Relative Humidity (%):"].get())
            cc = float(entries["Cloud Cover (%):"].get())

            if hm > 100 or cc > 100:
                messagebox.showwarning("Value Error", "Humidity & Cloud Cover metrics cannot physically scale past 100%.")
                return

            # Feature Scaling & Evaluation Transformation
            input_data = np.array([[rf, pr, hm, cc]])
            scaled_input = scaler.transform(input_data)

            if best_model_name == 'LSTM':
                scaled_input = scaled_input.reshape((1, 1, 4))
                prob = best_model.predict(scaled_input, verbose=0)[0][0]
            else:
                prob = best_model.predict_proba(scaled_input)[0][1]

            # Render Results UI Elements
            if prob >= 0.5:
                result_lbl.configure(foreground="#d9534f")
                result_var.set(f"⚠️ CLOUDBURST RISK DETECTED!\nProbability Vector Score: {prob:.2%}")
            else:
                result_lbl.configure(foreground="#5cb85c")
                result_var.set(f"✅ Environment Matrix Stable\nProbability Vector Score: {prob:.2%}")

        except ValueError:
            messagebox.showerror("Execution Error", "Please verify entries contain numeric float arrays.")

    # Action Evaluation Button
    calc_btn = ttk.Button(root, text="Evaluate Risk Profile", command=run_prediction)
    calc_btn.pack(pady=10)
    result_lbl.pack(pady=10)

# Execute the local application frame rendering loops
root.mainloop()
````
